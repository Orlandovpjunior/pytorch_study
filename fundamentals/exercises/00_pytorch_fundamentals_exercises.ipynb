### 2. Create a random tensor with shape `(7, 7)`.


````python
# Import torch
import torch

# Create random tensor
random_tensor = torch.rand(size=(7,7))
print(random_tensor, random_tensor.size(), random_tensor.dim())
````

### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 7)` (hint: you may have to transpose the second tensor).

````python
# Create another random tensor
another_random_tensor = torch.rand(size=(1,7))
# Perform matrix multiplication 
mult_matrix = torch.mm(random_tensor, another_random_tensor.T)
print(mult_matrix, mult_matrix.size())
````

### 4. Set the random seed to `0` and do 2 & 3 over again.

The output should be:
```
(tensor([[1.8542],
         [1.9611],
         [2.2884],
         [3.0481],
         [1.7067],
         [2.5290],
         [1.7989]]), torch.Size([7, 1]))
```

````python
# Set manual seed
SEED = 0
torch.manual_seed(SEED)
# Create two random tensors
new_tensor = torch.rand(size=(7,7))
new_another_tensor = torch.rand(size=(1,7))

# Matrix multiply tensors

new_mult = torch.mm(new_tensor,new_another_tensor.T)
print(new_mult, new_mult.size())
````

### 5. Speaking of random seeds, we saw how to set it with `torch.manual_seed()` but is there a GPU equivalent? (hint: you'll need to look into the documentation for `torch.cuda` for this one)
  * If there is, set the GPU random seed to `1234`.

````python
# Set random seed on the GPU
seed = 0
torch.cuda.random.manual_seed(seed)

# Create two random tensors
new_tensor_gpu = torch.rand(size=(7,7),device="cuda")
new_another_tensor_gpu = torch.rand(size=(1,7),device="cuda")

# Matrix multiply tensors

new_mul_gpu = torch.mm(new_tensor_gpu,new_another_tensor_gpu.T)
print(new_mul_gpu, new_mul_gpu.size())
````


### 6. Create two random tensors of shape `(2, 3)` and send them both to the GPU (you'll need access to a GPU for this). Set `torch.manual_seed(1234)` when creating the tensors (this doesn't have to be the GPU random seed). The output should be something like:

```
Device: cuda
(tensor([[0.0290, 0.4019, 0.2598],

 tensor([[0.0518, 0.4681, 0.6738],

```

````python
# Set random seed
rs = 42
torch.cuda.random.manual_seed(rs)
# Check for access to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Create two random tensors on GPU
tensor1 = torch.rand(size=(2,3),device=device)
tensor2 = torch.rand(size=(2,3),device=device)

print(f'Device: {device}')
print(tensor1)
print(tensor2)
````
