### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules that perfoming matrix multiplication needs to satisfy:

1. the **inner dimensions** must match:
* `(3,*2*) @ (*3*,2)` won't work
* `(2,*3*) @ (*3*,2)` will work
* `(3,*2*) @ (*2*,3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
* `(*2*,3) @ (3,*2*)` -> `(2,2)`
* `(*3*,2) @ (2,*3*)` -> `(3,3)`

````python
torch.matmul(torch.rand(2,3),torch.rand(3,2).shape)
````

````python
# Element wise multiplication

tensor = torch.tensor([1,2,3])
print(tensor,"*",tensor)
print(f'Equals: {tensor * tensor}')
````

### One of the most common eros in deep learning: shape error

To fix our tensor shape issues, we can manipulate the shape of one of our tensor using a ***transpose***

A transpose switches the axes or dimensios of a given tensor.

````python
# Shapes for matrix multiplication
tensor_A = torch.rand(size=(3,2))
tensor_B = torch.rand(size=(3,2))

torch.mm(tensor_A,tensor_B) # mm is the same as matmul
````

````python
tensor_B, tensor_B.shape
````

````python
tensor_B.T, tensor_B.shape
````

````python
# The matrix multiplication operation works when tensor_B is transposed

print(f'Originial shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}')
print(f'New shapes: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.t = {tensor_B.T.shape}')

torch.mm(tensor_A, tensor_B.T).shape
````

## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

````python
# Let's create a tensor
x = torch.arange(1.,10.)
x, x.shape
````

````python
# Add an extra dimension
x_reshaped = x.reshape(1,9) # the result of reshape(x,y) x*y must be the same as the arange(end - start)
x_reshaped, x_reshaped.shape
````

````python
# Change the view
z = x.view(1,9)
z, z.shape
````

````python
# Changing z changes x (because a view of a tensor shares the same memory as the original)
z[:,0] = 5
z,x
````

````python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
````

````python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim=1)
x_stacked
````

````python
# torch.squeeze() - remove all the single dimensions from a target tensor
print(f'Previus tensor: {x_reshaped}')
print(f'Previous shape: {x_reshaped.shape}')

x_squeezed = x_reshaped.squeeze()
print(f'\nNew tensor: {x_squeezed}')
print(f'New Shape: {x_squeezed.shape}')
````

````python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim
print(f'Previus target: {x_squeezed}')
print(f'Previus shape: {x_squeezed.shape}')

# Add an extra dimension with unsqueeze
x_unsqueeze = x_squeezed.unsqueeze(dim=0)
print(f'\nNew tensor: {x_unsqueeze}')
print(f'\nNew shape: {x_unsqueeze.shape}')
````

````python
# torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original= torch.rand(size=(224,224,3)) # [height, width, colour_channels]

# Permute the originial tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2,0,1)

print(f'Previous shape: {x_original.shape}')
print(f'New shape: {x_permuted.shape}')
````

````python
x_original[0,0,0] = 728218
x_original[0,0,0], x_permuted[0,0,0]
````

## Pytorch & Numpy

numpy is a popular scientific Python numerical computing library.
And because of this, pytorch has functionality to interact with it.

* Data in Numpy, want in Pytorch tensor -> `torch.from_numpy(ndarray)`
* Pytorch tensor -> Numpy -> `torch.tensor.numpy()`

````python
import numpy as np
import torch

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array) # warning: when converting from numpy -> pytorch reflects numpy's default datatype of float 64 unless specified otherwise
array,tensor
````

````python
# Change the value of array, what will this do to `tensor`?
array = array + 1
array, tensor
````

````python
# Tensor to Numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor,numpy_tensor # default float32, Be careful
````

````python
# Change the value of tensor, what will this do to `numpy_tensor`?
tensor = tensor + 1
tensor, numpy_tensor
````

## Reproducbility (trying to take random out of random)

In short how a neural network learns:
 `start with random numhers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> ...`

To reduce the randomness in neural networks and Pytorch comes the concept of **random seed**

Essentially what the random seed does is "flavour" the randomness.

````python
import torch

random_tensor_A = torch.rand(size=(3,4))
random_tensor_B = torch.rand(size=(3,4))

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
````

````python
# Let's make some random but reproducible tensors
# Set the random seed

RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_c = torch.rand(3,4) # You have to put the manual seed for each tensor
torch.manual_seed(RANDOM_SEED)
random_tensor_d = torch.rand(3,4)

print(random_tensor_c)
print(random_tensor_d)
print(random_tensor_c == random_tensor_d)
````
